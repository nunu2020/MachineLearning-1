# 第一题  （参考 2019202193-吕良波）

&emsp;&emsp;从香农对于信息的熵的定义来看，当我们所具有的信息越多的时候，信息所对应的混乱程度也就是信息熵要越小。由于条件概率所知道的信息情况是多于普通的概率的，因此可以认为条件熵是小于原本的熵的。这个是这个不等式字面上所要传达的性质。下面我们就对该不等式进行数学上的证明。

&emsp;&emsp;我们主要对条件熵的积分形式进行证明。离散的情况是求和形式，可以看作一种特殊的积分形式，具体的证明过程是一样的。我们就在这里不进行讨论。我们带入信息熵的计算式并利用对数函数是凸函数的性质可得:

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$H(x|y)=-\int \int p(x,y)logp(x|y)dxdy$
&emsp;&emsp;&emsp;&emsp;$H(x)=-\int \int p(x,y)logp(x)dxdy$

&emsp;$H(x)-H(x|y)=-Elog(\frac{p(x)p(y)}{p(x,y)}) \geq -log(E\frac{p(x)p(y)}{p(x,y)}) = -log(\int \int p(x,y) \frac{p(x)p(y)}{p(x,y)}dxdy) =-log( \int p(x)dx \int p(y)dy) = -log(1)=0$

&emsp;&emsp;由上我们可以得到最终的结论:$H(x) \geq H(x|y)$并可以看到，当$p(x)p(y)=p(x,y)$也就是x,y相互独立的时候取等。

# 第二题 （参考 2019202091-徐克丰）
## 题目描述
计算敲声的信息增益和信息增益比率，通过例题分析由信息增益相同时所导致的对不同子树的随机选择，在信息增益比率的之下是否可以避免，给出算法和分析讨论。
## 结果分析
(1)计算“敲声”的信息增益和信息增益比率

属性“敲声”有3个取值$(V=3)$，分别为“浊响”，对应样例$D^1=\{1,3,5,6,7,8,12,13,15,16\}$；“沉闷”，对应样例$D^2=\{2,4,9,14,17\}$；“清脆”,对应样例$D^3=\{10,11\}$。分别计算这三个样例集合的熵：
$$\begin{eqnarray*}Ent(D^1)&=&-(\frac{6}{10}log_2\frac{6}{10}+\frac{4}{10}log_2\frac{4}{10})=0.971\\
Ent(D^2)&=&-(\frac{2}{5}log_2\frac{2}{5}+\frac{3}{5}log_2\frac{3}{5})=0.971\\
Ent(D^3)&=&-(\frac{0}{2}log_2\frac{0}{2}+\frac{2}{2}log_2\frac{2}{2})=0\end{eqnarray*}$$

计算敲声的信息增益的过程为：
$$\begin{eqnarray*}Ent(D)&=&-\sum_{k=1}^{2}p_klog_2p_k=-(\frac{8}{17}log_2\frac{8}{17}+\frac{9}{17}log_2\frac{9}{17})=0.998\\
Gain(D，敲声)&=&Ent(D)-\sum_{v=1}^{3}\frac{|D^v|}{|D|}Ent(D^v)=0.998-(\frac{10}{17}\times{0.971}+\frac{5}{17}\times{0.971})=0.141\end{eqnarray*}$$

计算敲声信息增益比率的过程为：
$$\begin{eqnarray*}IV(a)=-\sum_{v=1}^{3}\frac{|D^v|}{|D|}log_2\frac{|D^v|}{|D|}&=&-(\frac{10}{17}log_2\frac{10}{17}+\frac{5}{17}log_2\frac{5}{17}+\frac{2}{17}log_2\frac{2}{17})=1.333\\
Gain\_ratio(D,a)&=&\frac{Gain(D,a)}{IV(a)}=\frac{0.141}{1.333}=0.106\end{eqnarray*}$$

(2)分析信息增益相同的情况下采用信息增益比率可否避免随机选择

在PPT17页中，假设已经完成了以“纹理”为属性的一次划分，而得到的“纹理=清晰”的集合为$D^1=\{1,2,3,4,5,6,8,10,15\}$。则此时有三个属性均取得了最大的信息增益，分别为：$$Gain(D^1,根蒂)=Gain(D^1,脐部)=Gain(D^1,触感)=0.458$$
这时分别计算三种属性的固有值：
$$\begin{eqnarray*}IV(a=根蒂)=-\sum_{v=1}^{3}\frac{|D^{1v}|}{|D^1|}log_2\frac{|D^{1v}|}{|D^1|}&=&-(\frac{5}{9}log_2\frac{5}{9}+\frac{3}{9}log_2\frac{3}{9}+\frac{1}{9}log_2\frac{1}{9})=1.352 \\
IV(a=脐部)=-\sum_{v=1}^{3}\frac{|D^{1v}|}{|D^1|}log_2\frac{|D^{1v}|}{|D^1|}&=&-(\frac{5}{9}log_2\frac{5}{9}+\frac{3}{9}log_2\frac{3}{9}+\frac{1}{9}log_2\frac{1}{9})=1.352\\
IV(a=触感)=-\sum_{v=1}^{2}\frac{|D^{1v}|}{|D^1|}log_2\frac{|D^{1v}|}{|D^1|}&=&-(\frac{6}{9}log_2\frac{6}{9}+\frac{3}{9}log_2\frac{3}{9})=0.918\\\end{eqnarray*}$$
以上述过程为例，三个属性根蒂，脐部和触感的信息增益相同，则通过计算三者的固有值即可比较出三者的信息增益率，固有值越小，信息增益率越大。发现$IV(a=触感)$=0.918小于其他两个属性的固有值，所以对于本例题而言，由于信息增益相同所导致的对不同子树的随机选择可以避免。事实上，增益率对可取值数目较少的属性有所偏好，算法中常用的并不是直接选择增益率最大的候选划分属性，而是先从候选划分属性中找出信息增益高于平均水平的属性，再从中选择增益率最高的。

```python
#引入给出的决策树生成函数，以参考示例代码为基础，添加注释
#结合题目要求(生成两层决策树)，对决策树的生成进行了调整，重写了决策树分类函数
from math import log
import operator
import os

import re
from numpy import inf
import copy


#计算某个特征中未缺失该特征样本的信息熵
def calcShannonEnt(dataSet, labelIndex): 
    numEntries = 0  # 样本数(按权重计算）
    labelCounts = {} #字典类型，储存特征的取值和取不同值样本的数目
    for featVec in dataSet:  # 遍历每个样本，不计入有缺失样本
        if featVec[labelIndex] != 'N':   
            weight = float(featVec[-2]) #取该样本的权重
            numEntries += weight
            currentLabel = featVec[-1]  # 当前样本的类别,Play or Don't Play
            if currentLabel not in labelCounts.keys():  # 生成类别字典,并设置初值为0
                labelCounts[currentLabel] = 0
            labelCounts[currentLabel] += weight  # 数据集的倒数第二个值用来标记样本权重，计算出某一类别的权重
    shannonEnt = 0.0
    for key in labelCounts:  # 计算信息熵，对于所有类别遍历，这里最多有两类
        prob = float(labelCounts[key]) / numEntries #计算pk，即第k类的权重除以未缺失样本总权重之和
        shannonEnt = shannonEnt - prob * log(prob, 2) #循环计算每一类的pklog2k并求和
    return shannonEnt

#划分数据集，不划分在某个特征上缺失的样本，按照某个特征的某个取值提取出一部分样本
def splitDataSet(dataSet, axis, value, LorR='N'):  
    """
    type: (list, int, string or float, string) -> list
    划分数据集
    axis:按第几个特征划分
    value:划分特征的值
    LorR: N 离散属性; L 小于等于value值; R 大于value值
    """
    retDataSet = []
    featVec = []  
    if LorR == 'N':  #对于属性是离散的情况
        for featVec in dataSet:
            if featVec[axis] == value:  #如果特征向量与值相等，这里就自然排除了属性缺失的情况
                reducedFeatVec = featVec[:axis] #就取0,..i-1,i+1,...的属性不包括划分特征
                reducedFeatVec.extend(featVec[axis + 1:])
                retDataSet.append(reducedFeatVec) #取该特征等于某个值的样本列表，并删去该特征这一列
    elif LorR == 'L': #提取在某连续属性上小于阈值的数据集合
        for featVec in dataSet:
            if featVec[axis] != 'N': #排除了连续属性样本缺失的情况
                if float(featVec[axis]) < value: 
                    retDataSet.append(featVec) #D-,注意这里没有减少特征的数目
    elif LorR == 'R': #提取在某连续属性上大于阈值的数据集合
        for featVec in dataSet:
            if featVec[axis] != 'N':
                if float(featVec[axis]) > value: #取大于阈值的D+
                    retDataSet.append(featVec)
    return retDataSet

#划分数据集，对于某属性上缺失的样本，按照概率分配不同权重到各个子节点中
def splitDataSetWithNull(dataSet, axis, value, LorR='N'):  
    """
    type: (list, int, string or float, string) -> list
    划分数据集
    axis:按第几个特征划分
    value:划分特征的值
    LorR: N 离散属性; L 小于等于value值; R 大于value值
    """
    retDataSet = [] #储存正常的没有缺失值的样本
    nullDataSet = [] #储存带有缺失值的样本
    featVec = [] 
    totalWeightV = calcTotalWeight(dataSet, axis, False)  # 非空样本权重
    totalWeightSub = 0.0
    if LorR == 'N':  # 离散属性
        for featVec in dataSet:
            if featVec[axis] == value: #属性取值value的样本，减少一个特征后划入
                reducedFeatVec = featVec[:axis]
                reducedFeatVec.extend(featVec[axis + 1:])
                retDataSet.append(reducedFeatVec)
            elif featVec[axis] == 'N': #属性缺失的样本，划入缺失集合
                reducedNullVec = featVec[:axis]
                reducedNullVec.extend(featVec[axis + 1:])
                nullDataSet.append(reducedNullVec)
    elif LorR == 'L':   #连续属性，D-
        for featVec in dataSet:
            if featVec[axis] != 'N': #属性不缺失的样本，小于阈值的划入到列表中
                if float(featVec[axis]) < value:
                    retDataSet.append(featVec)
            elif featVec[axis] == 'N': #属性缺失的划入一个列表中
                nullDataSet.append(featVec)
    elif LorR == 'R':
        for featVec in dataSet:
            if featVec[axis] != 'N': #属性不缺失的样本，大于阈值的划入到列表中
                if float(featVec[axis]) > value:
                    retDataSet.append(featVec)
            elif featVec[axis] == 'N': #属性缺失的划入一个列表中
                nullDataSet.append(featVec)

    totalWeightSub = calcTotalWeight(retDataSet, -1, True)  #计算此分支中非空样本的总权重
    for nullVec in nullDataSet:  # 把缺失值样本按权值比例划分到分支中
        nullVec[-2] = float(nullVec[-2]) * totalWeightSub / totalWeightV  #计算rv，修正其权重值
        retDataSet.append(nullVec)  #将修改权重后的集合合并，返回一个带有权重的属性划分子集

    return retDataSet

#按照某一属性计算该属性上的权重，分为样本属性中包含NULL和不包含NULL的情况
def calcTotalWeight(dataSet, labelIndex, isContainNull):
    """
    type: (list, int, bool) -> float
    计算样本集对某个特征值的总样本树（按权重计算）
    :param dataSet: 数据集
    :param labelIndex: 特征值索引
    :param isContainNull: 是否包含空值的样本
    :return: 返回样本集的总权重值
    """
    totalWeight = 0.0
    for featVec in dataSet:  # 遍历每个样本
        weight = float(featVec[-2])
        if isContainNull is False and featVec[labelIndex] != 'N':
            totalWeight += weight  # 非空样本树，按权重计算。权重不包括缺失值时，加和不缺失样本的权重
        if isContainNull is True:
            totalWeight += weight  #如果包括缺失，则包括缺失点一起加
    return totalWeight

#计算某一属性的信息增益，计算信息增益不考虑在该属性上缺失的样本
def calcGain(dataSet, labelIndex, labelPropertyi):  
    """
    type: (list, int, int) -> float, int
    计算信息增益,返回信息增益值和连续属性的划分点
    dataSet: 数据集
    labelIndex: 特征值索引
    labelPropertyi: 特征值类型，0为离散，1为连续
    """
    baseEntropy = calcShannonEnt(dataSet, labelIndex)  # 计算根节点的信息熵
    #注意计算信息增益而选择数据划分时，只用该属性不缺失的样本划分即可
    featList = [example[labelIndex] for example in dataSet]  # 特征值列表，只考虑特征值一列
    uniqueVals = set(featList)  # 该特征包含的所有值，取集合，表示某个属性的不同取值
    newEntropy = 0.0 #新的信息熵
    totalWeight = 0.0
    totalWeightV = 0.0
    totalWeight = calcTotalWeight(dataSet, labelIndex, True)  # 总样本权重
    totalWeightV = calcTotalWeight(dataSet, labelIndex, False)  # 非空样本权重，这两行相除得出密度rou
    if labelPropertyi == 0:  # 对离散的特征
        for value in uniqueVals:  # 对每个不缺失的特征值，划分数据集, 计算各子集的信息熵
            if value != 'N':  
                subDataSet = splitDataSet(dataSet, labelIndex, value) #划分无缺失样本，特定属性==value
                totalWeightSub = 0.0
                totalWeightSub = calcTotalWeight(subDataSet, labelIndex, True) #计算出==value的无缺失样本
                prob = totalWeightSub / totalWeightV #计算rv，无缺失样本值取值value的比例
                newEntropy += prob * calcShannonEnt(subDataSet, labelIndex) #计算信息熵
    else:  # 对连续的特征
        uniqueValsList = list(uniqueVals)   #取出现过的所有取值
        if 'N' in uniqueValsList:
            uniqueValsList.remove('N')  #剔除缺失值
        sortedUniqueVals = sorted(uniqueValsList)  # 对特征值排序
        listPartition = []  #列表划分
        minEntropy = inf   #inf是最小值，即这里需要找最小的熵
        if len(sortedUniqueVals) == 1:  # 如果只有一个值，可以看作只有左子集，没有右子集
            totalWeightLeft = calcTotalWeight(dataSet, labelIndex, True)  #计算D-的权重
            probLeft = totalWeightLeft / totalWeightV  #计算rv，无缺样本取D-的比例
            minEntropy = probLeft * calcShannonEnt(dataSet, labelIndex) #则最小熵对应rv*ent(Dv)
        else:
            for j in range(len(sortedUniqueVals) - 1):  # 计算划分点
                partValue = (float(sortedUniqueVals[j]) + float(
                    sortedUniqueVals[j + 1])) / 2    #得出划分点候选集合Ta中的一个值
                # 对每个划分点，计算信息熵
                dataSetLeft = splitDataSet(dataSet, labelIndex, partValue, 'L') #划分无缺失样本，D-
                dataSetRight = splitDataSet(dataSet, labelIndex, partValue, 'R') #划分无缺时样本，D+
                totalWeightLeft = 0.0
                totalWeightLeft = calcTotalWeight(dataSetLeft, labelIndex, True) 
                totalWeightRight = 0.0
                totalWeightRight = calcTotalWeight(dataSetRight, labelIndex, True) #分别计算D-、D+的集合权重
                probLeft = totalWeightLeft / totalWeightV  #分别计算rv
                probRight = totalWeightRight / totalWeightV
                Entropy = probLeft * calcShannonEnt(dataSetLeft, labelIndex) + \
                          probRight * calcShannonEnt(dataSetRight, labelIndex)  #计算信息熵
                if Entropy < minEntropy:  # 取最小的信息熵，可以使得信息增益最大
                    minEntropy = Entropy
        newEntropy = minEntropy
    gain = totalWeightV / totalWeight * (baseEntropy - newEntropy) #信息增益再乘以\rou，得到Gain(D,a)
    return gain #这里可以再添加一个返回值，Partvalue

#计算某一属性的信息增益率
def calcGainRatio(dataSet, labelIndex, labelPropertyi):
    """
    type: (list, int, int) -> float, int
    计算信息增益率,返回信息增益率和连续属性的划分点
    dataSet: 数据集
    labelIndex: 特征值索引
    labelPropertyi: 特征值类型，0为离散，1为连续
    """
    baseEntropy = calcShannonEnt(dataSet, labelIndex)  # 计算根节点的信息熵
    featList = [example[labelIndex] for example in dataSet]  # 特征值列表
    uniqueVals = set(featList)  # 该特征包含的所有值
    newEntropy = 0.0
    bestPartValuei = None
    IV = 0.0
    totalWeight = 0.0
    totalWeightV = 0.0
    totalWeight = calcTotalWeight(dataSet, labelIndex, True)  # 总样本权重
    totalWeightV = calcTotalWeight(dataSet, labelIndex, False)  # 非空样本权重
    if labelPropertyi == 0:  # 对离散的特征
        for value in uniqueVals:  # 对每个特征值，划分数据集, 计算各子集的信息熵
            subDataSet = splitDataSet(dataSet, labelIndex, value)
            totalWeightSub = 0.0
            totalWeightSub = calcTotalWeight(subDataSet, labelIndex, True)
            if value != 'N':
                prob = totalWeightSub / totalWeightV
                newEntropy += prob * calcShannonEnt(subDataSet, labelIndex)
            prob1 = totalWeightSub / totalWeight
            IV -= prob1 * log(prob1, 2) #计算属性的固有值
    else:  # 对连续的特征
        uniqueValsList = list(uniqueVals)
        if 'N' in uniqueValsList:
            uniqueValsList.remove('N')
            # 计算空值样本的总权重，用于计算IV
            totalWeightN = 0.0
            dataSetNull = splitDataSet(dataSet, labelIndex, 'N') #按照离散属性划分样本，不包括空值
            totalWeightN = calcTotalWeight(dataSetNull, labelIndex, True) 
            probNull = totalWeightN / totalWeight
            if probNull > 0.0:
                IV += -1 * probNull * log(probNull, 2)  

        sortedUniqueVals = sorted(uniqueValsList)  # 对特征值排序
        listPartition = []
        minEntropy = inf

        if len(sortedUniqueVals) == 1:  # 如果只有一个值，可以看作只有左子集，没有右子集
            totalWeightLeft = calcTotalWeight(dataSet, labelIndex, True)
            probLeft = totalWeightLeft / totalWeightV
            minEntropy = probLeft * calcShannonEnt(dataSet, labelIndex)
            IV = -1 * probLeft * log(probLeft, 2)
        else:
            for j in range(len(sortedUniqueVals) - 1):  # 计算划分点
                partValue = (float(sortedUniqueVals[j]) + float(
                    sortedUniqueVals[j + 1])) / 2
                # 对每个划分点，计算信息熵
                dataSetLeft = splitDataSet(dataSet, labelIndex, partValue, 'L')
                dataSetRight = splitDataSet(dataSet, labelIndex, partValue, 'R')
                totalWeightLeft = 0.0
                totalWeightLeft = calcTotalWeight(dataSetLeft, labelIndex, True)
                totalWeightRight = 0.0
                totalWeightRight = calcTotalWeight(dataSetRight, labelIndex, True)
                probLeft = totalWeightLeft / totalWeightV
                probRight = totalWeightRight / totalWeightV
                Entropy = probLeft * calcShannonEnt(
                    dataSetLeft, labelIndex) + probRight * calcShannonEnt(dataSetRight, labelIndex)
                if Entropy < minEntropy:  # 取最小的信息熵
                    minEntropy = Entropy
                    bestPartValuei = partValue
                    probLeft1 = totalWeightLeft / totalWeight
                    probRight1 = totalWeightRight / totalWeight
                    IV += -1 * (probLeft1 * log(probLeft1, 2) + probRight1 * log(probRight1, 2))

        newEntropy = minEntropy
    gain = totalWeightV / totalWeight * (baseEntropy - newEntropy)
    if IV == 0.0:  # 如果属性只有一个值，IV为0，为避免除数为0，给个很小的值
        IV = 0.0000000001
    gainRatio = gain / IV  
    return gainRatio, bestPartValuei #返回信息增益率合最佳的划分阈值

#选择最好的数据集划分方式，返回最佳划分属性的索引和连续属性的划分值，labelProperty是所有属性的类型(连续/离散)
def chooseBestFeatureToSplit(dataSet, labelProperty):
    """
    type: (list, int) -> int, float
    :param dataSet: 样本集
    :param labelProperty: 特征值类型，1 连续， 0 离散
    :return: 最佳划分属性的索引和连续属性的划分值
    """
    numFeatures = len(labelProperty)  # 特征数
    bestInfoGainRatio = 0.0
    bestFeature = -1
    bestPartValue = None  # 连续的特征值，最佳划分值
    gainSum = 0.0
    gainAvg = 0.0
    for i in range(numFeatures):  # 对每个特征循环，计算信息增益
        infoGain = calcGain(dataSet, i, labelProperty[i]) #表明特征值类型，以计算信息增益均值
        gainSum += infoGain
    gainAvg = gainSum / numFeatures
    for i in range(numFeatures):  # 对每个特征循环
        infoGainRatio, bestPartValuei = calcGainRatio(dataSet, i, labelProperty[i])  #计算特征的最大信息增益，bestPartValuei对于离散型返回None
        infoGain = calcGain(dataSet, i, labelProperty[i]) #获取每个特征的信息增益
        if infoGainRatio > bestInfoGainRatio and infoGain > gainAvg:  # 取信息增益高于平均增益且信息增益率最大的特征
            bestInfoGainRatio = infoGainRatio
            bestFeature = i
            bestPartValue = bestPartValuei
    return bestFeature, bestPartValue  


# 通过排序返回出现次数最多的类别，以权重衡量
def majorityCnt(classList, weightList):
    classCount = {}
    for i in range(len(classList)): #对于每一个样本进行遍历，从而计算每一类的权重累积(或者样本个数)
        if classList[i] not in classCount.keys(): 
            classCount[classList[i]] = 0.0 #对于未识别的类别，添加到字典中
        classCount[classList[i]] += round(float(weightList[i]),1)  #四舍五入的权重，1表示1位小数(在不含缺失样本时，就是样本个数)

    sortedClassCount = sorted(classCount.items(),                  #sorted排序，以每一类的值(每一类的样本个数)从大到小排序
                              key=operator.itemgetter(1), reverse=True) 
    if len(sortedClassCount) == 1:   #如果仅有一个类别
        return (sortedClassCount[0][0],sortedClassCount[0][1],0.0) #返回其类别，该类别的样本个数(或权重),0
    return (sortedClassCount[0][0], sortedClassCount[0][1], sortedClassCount[1][1]) #返回类别，该类别权重，样本集中另一类的个数(权重)


# 创建树, 样本集 特征 特征属性（0 离散， 1 连续），限制决策树最多有两层，即两层非叶节点，超出的部分全部记为叶节点
def createTree(dataSet, labels, labelProperty,deep=1): #;ables是属性集合，labelProperty包含了所有属性的类型(离散还是连续)
    classList = [example[-1] for example in dataSet]  # 训练集每个样本类别构成的向量，Play OR Don't Play
    weightList = [example[-2] for example in dataSet]  # 权重向量，包含了每个训练样本的初始权重，应设为1
    if classList.count(classList[0]) == len(classList):  # 如果只有一个类别,返回元组
        totalWeiht = calcTotalWeight(dataSet,0,True)   
        return (classList[0], round(totalWeiht,1),0.0) #设为叶节点，返回类别，DataSet中该类的总权重(样本个数)，另一类的样本个数
    if len(dataSet[0]) == 1:  #如果所有特征都被遍历完了(每筛选一个特征，样本特征就-1)，返回出现次数最多的类别
        return majorityCnt(classList,weightList)     #长度为1就是仅剩了weight这一列  
    if deep==3: #第三层的数据集只能是叶子结点，直接返回出现次数最多的类别
        return majorityCnt(classList, weightList)
    bestFeat,bestPartValue = chooseBestFeatureToSplit(dataSet,
                                                       labelProperty)  # 最优分类特征的索引和划分值
    if bestFeat == -1:  # 如果无法选出最优分类特征，返回出现次数最多的类别
        return majorityCnt(classList, weightList)
    if labelProperty[bestFeat] == 0:  # 若最优特征是离散的特征
        bestFeatLabel = labels[bestFeat] #最好索引对应的特征，是一个字符串属性标签
        myTree = {bestFeatLabel: {}} #用字典来储存决策树
        labelsNew = copy.copy(labels) #复制特征集合
        labelPropertyNew = copy.copy(labelProperty) #复制特征属性集合
        del (labelsNew[bestFeat])  # 已经选择的特征不再参与分类
        del (labelPropertyNew[bestFeat])
        featValues = [example[bestFeat] for example in dataSet] #选取数据集中的最好特征
        uniqueValue = set(featValues)  # 包含了该特征的所有取值
        uniqueValue.discard('N')   #扔掉缺失值标签
        for value in uniqueValue:  #对每个特征的取值，递归构建决策树
            subLabels = labelsNew[:] 
            subLabelProperty = labelPropertyNew[:]
            myTree[bestFeatLabel][value] = createTree(
                splitDataSetWithNull(dataSet, bestFeat, value), subLabels,
                subLabelProperty,deep+1)  #用符合特征值且包括特征缺失的样本构建子树
    else:  #对连续的特征，不删除该特征，分别构建左子树和右子树
        bestFeatLabel = labels[bestFeat] + '<' + str(bestPartValue) #设置内部节点连续属性的判断标签
        myTree = {bestFeatLabel: {}} #用字典来储存决策树
        subLabels = labels[:]   #深拷贝标签
        subLabelProperty = labelProperty[:]
        # 构建左子树
        valueLeft = 'Y' #构建复合映射{'bestfeatlabel':{'Y':subtree}}
        myTree[bestFeatLabel][valueLeft] = createTree(
            splitDataSetWithNull(dataSet, bestFeat, bestPartValue, 'L'), subLabels,
            subLabelProperty,deep+1)
        # 构建右子树
        valueRight = 'N'
        myTree[bestFeatLabel][valueRight] = createTree(
            splitDataSetWithNull(dataSet, bestFeat, bestPartValue, 'R'), subLabels,
            subLabelProperty,deep+1)
    return myTree

#设置决策树的分类，其基本原理是树的遍历，用递归方式，不考虑测试集合中的缺失样本
def classify(inputTree, classList, featLabels, featLabelProperties, testVec):
    if type(inputTree).__name__=='tuple':  #如果是元组，表示传入了叶子结点
        #tup=inputTree
        return inputTree #返回这个元组
    #如果不是元组，就是一颗树，则进行决策
    firstStr = list(inputTree.keys())[0]  #根节点对应的标签,如Temp<80,或者叶子结点对应的元组{'N': ("Don't Play", 1.0, 0.0)}
    lessIndex = str(firstStr).find('<')  #返回字符串索引
    if lessIndex > -1:  # 如果是连续型的特征
        firstLabel = str(firstStr)[:lessIndex] #示例：取Temp<80中的Temp
        laststr=str(firstStr)[lessIndex+1:]  #取出分类阈值
    else:
        firstLabel=firstStr
    secondDict = inputTree[firstStr]   #获取key=firstStr(比如Temp<80)对应的值(是一个字典或者元组)
    featIndex = featLabels.index(firstLabel)  # 提取比如label Temp在featLeabels中对应的索引值
    #print(firstLabel)
    if lessIndex==-1:  #对于离散型特征 
        for key in secondDict.keys(): #否则就根据分支寻找叶节点，比如连续属性的'Y'或'N'
            if str(testVec[featIndex])==str(key): #如果测试样例属性与某一分支相等，就进入该分支
                tup=classify1(secondDict[key], classList, featLabels, featLabelProperties, testVec)
    else:  #对于连续型特征 
            if testVec[featIndex]<eval(laststr): #如果测试样例连续属性小于阈值
                tup=classify1(secondDict['Y'], classList, featLabels, featLabelProperties, testVec)
            if testVec[featIndex]>=eval(laststr): #如果测试样例连续属性大于等于阈值
                tup=classify1(secondDict['N'], classList, featLabels, featLabelProperties, testVec)
    return tup  #返回最终的决策的元组
```

```python
#用示例的代码直接导入的决策树绘制代码块
import matplotlib.pyplot as plt

# 设置决策节点和叶节点的边框形状、边距和透明度，以及箭头的形状
decisionNode = dict(boxstyle="square,pad=0.5", fc="0.9")
leafNode = dict(boxstyle="round4, pad=0.5", fc="0.9")
arrow_args = dict(arrowstyle="<-", connectionstyle="arc3", shrinkA=0,
                  shrinkB=16)


def plotNode(nodeTxt, centerPt, parentPt, nodeType):
    newTxt = nodeTxt
    if type(nodeTxt).__name__ == 'tuple':
        newTxt = nodeTxt[0] + '\n'
        for strI in nodeTxt[1:-1]:
            newTxt += str(strI) + ','
        newTxt+= str(nodeTxt[-1])

    createPlot.ax1.annotate(newTxt, xy=parentPt,
                            xycoords='axes fraction',
                            xytext=centerPt, textcoords='axes fraction',
                            va="top", ha="center", bbox=nodeType,  #bbox=nodeType
                            arrowprops=arrow_args)


def getNumLeafs(myTree):
    numLeafs = 0
    firstStr = list(myTree.keys())[0]
    secondDict = myTree[firstStr]
    for key in secondDict.keys():
        if type(secondDict[key]).__name__ == 'dict':
            numLeafs += getNumLeafs(secondDict[key])
        else:
            numLeafs += 1
    return numLeafs


def getTreeDepth(myTree):
    maxDepth = 0
    firstStr = list(myTree.keys())[0]
    secondDict = myTree[firstStr]
    for key in secondDict.keys():
        if type(secondDict[key]).__name__ == 'dict':
            thisDepth = 1 + getTreeDepth(secondDict[key])
        else:
            thisDepth = 1
        if thisDepth > maxDepth: maxDepth = thisDepth
    return maxDepth


def retrieveTree(i):
    listOfTrees = [
        {'no surfacing': {0: 'no', 1: {'flippers': {0: 'no', 1: 'yes'}}}},
        {'no surfacing': {0: 'no', 1: {
            'flippers': {0: {'head': {0: 'no', 1: 'yes'}}, 1: 'no'}}}}
    ]
    return listOfTrees[i]


def plotMidText(cntrPt, parentPt, txtString):
    xMid = (parentPt[0] - cntrPt[0]) / 2.0 + cntrPt[0]
    yMid = (parentPt[1] - cntrPt[1]) / 2.0 + cntrPt[1]
    createPlot.ax1.text(xMid, yMid, txtString)


def plotTree(myTree, parentPt, nodeTxt):
    numLeafs = getNumLeafs(myTree)
    depth = getTreeDepth(myTree)
    firstStr = list(myTree.keys())[0]
    cntrPt = (plotTree.xOff + (1 + float(numLeafs)) / 2.0 / plotTree.totalW,
              plotTree.yOff)
    plotMidText(cntrPt, parentPt, nodeTxt)
    plotNode(firstStr, cntrPt, parentPt, decisionNode)
    secondDict = myTree[firstStr]
    plotTree.yOff = plotTree.yOff - 1.0 / plotTree.totalD
    for key in secondDict.keys():
        if type(secondDict[key]).__name__ == 'dict':
            plotTree(secondDict[key], cntrPt, str(key))
        else:
            plotTree.xOff = plotTree.xOff + 1.0 / plotTree.totalW
            plotNode(secondDict[key], (plotTree.xOff, plotTree.yOff),
                     cntrPt, leafNode)
            plotMidText((plotTree.xOff, plotTree.yOff), cntrPt, str(key))
    plotTree.yOff = plotTree.yOff + 1.0 / plotTree.totalD


def createPlot(inTree):
    fig = plt.figure(1, facecolor='white')
    fig.clf()
    axprops = dict(xticks=[], yticks=[])
    createPlot.ax1 = plt.subplot(111, frameon=False, **axprops)
    plotTree.totalW = float(getNumLeafs(inTree))
    plotTree.totalD = float(getTreeDepth(inTree)) + 0.5
    plotTree.xOff = -0.5 / plotTree.totalW
    plotTree.yOff = 1.0
    plotTree(inTree, (0.5, 1.0), '')
    plt.show()
```
